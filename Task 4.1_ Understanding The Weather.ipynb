```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

Import the weather chart.

```python
x=pd.read_csv('/content/drive/MyDrive/machine/Weather Dataset.csv')
print(x)
```

step to check there is NaN value in the data set

```python
print(x.isnull())
print(x.isnull().sum())
```

check if there are duplicates in the dataset to remove it.

```python
x2=x.drop_duplicates(subset='Formatted Date',keep='first')
print(x.shape)
print(x2.shape)
```

Plot temperature over time.

```python
x2.plot(kind = 'line', x = 'Formatted Date', y = 'Apparent Temperature (C)',xlabel='Time',ylabel='Temperature (C)')
plt.show()
```

Create a histogram of temperature distribution.

```python
x2.hist(column='Apparent Temperature (C)')
```

Plot a scatter plot of temperature vs. humidity.

```python
x2.plot.scatter(x='Humidity',y='Apparent Temperature (C)',ylabel="Temperature (C)",s=10)
plt.show()
```

Generate a correlation heatmap of the dataset

```python
heat_map=x2.drop(columns=['Formatted Date','Precip Type','Summary','Daily Summary'])
sns.heatmap(heat_map.corr())
plt.show()
```
